```python
# Your code here...
#1 Initialize KNN classifier using KNeighborsClassifier()
knn = KNeighborsClassifier()
knn
```

```python
#Define parameters for n_neighbors ranging from 1 to 50
param_grid = {
    'n_neighbors': np.arange(1, 51)}
```

```python
#implement a grid search ursing GridSearchCV with 10-fold cross-validation
grid = GridSearchCV(knn, param_grid, cv=10)
```
